```python
from bs4 import BeautifulSoup
from splinter import Browser
```

```python
# create instance
browser = Browser("chrome")

url ="https://finance.yahoo.com/losers/"
browser.visit(url)

html = browser.html
html
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
# target elements
target = soup.find("table",class_="W(100%)")
allstock_block = target.find_all("tr", class_="simpTblRow")

# Stock list
stock_list = []
percentage_change_list=[]

for i in range(len(allstock_block)):
    single_stock_block = allstock_block[i]

    # get stock name
    stock_symbol = single_stock_block.find("a", class_="Fw(600) C($linkColor)")

    # get stock % change
    find_percentage_change = single_stock_block.find("td", class_="Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)", attrs={"aria-label": "% Change"})
    percentage_change = find_percentage_change.find("span",class_="C($negativeColor)")


    # append to list
    stock_list.append(stock_symbol.text)
    percentage_change_list.append(percentage_change.text)
```

```python
zipped_list = list(zip(stock_list, percentage_change_list))
print(zipped_list)
```

```python
import csv
import datetime

# Get today's date
today_date = datetime.date.today()

# Create CSV file name
csv_file_name = f"top_loser_stock_{today_date}.csv"

# Define headers
headers = ["Symbol", "Percentage Change"]

# Export to CSV
with open(csv_file_name, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # Write headers to the CSV file
    writer.writerow(headers)

    # Write each row to the CSV file
    for row in zipped_list:
        writer.writerow(row)

print("CSV file created successfully:", csv_file_name)
```
